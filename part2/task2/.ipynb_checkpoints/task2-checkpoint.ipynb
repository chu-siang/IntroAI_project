# Part 2 Task 2 — Modeling & Enhancements

This notebook section is for Sofi's part: baseline TF-IDF+SVM, advanced model comparison, and Kaggle submission generation.

```python
# If needed in Colab:
# !pip install xgboost -q

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, classification_report

from xgboost import XGBClassifier

RANDOM_STATE = 42
```

```python
train_df = pd.read_csv("train.csv").rename(columns={"content": "text"})
test_df = pd.read_csv("test.csv").rename(columns={"content": "text"})

train_df["text"] = (
    train_df["text"]
    .astype(str)
    .str.lower()
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

test_df["text"] = (
    test_df["text"]
    .astype(str)
    .str.lower()
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

print("Train:", train_df.shape)
print("Test:", test_df.shape)

print("Label distribution:")
print(train_df["label"].value_counts())

display(train_df.head())
```

```python
X = train_df["text"]
y = train_df["label"].astype(int)

X_train, X_val, y_train, y_val = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=RANDOM_STATE
)

print("Train labels:")
print(y_train.value_counts())

print("Validation labels:")
print(y_val.value_counts())
```

```python
baseline_svm = Pipeline([
    ("tfidf", TfidfVectorizer(
        ngram_range=(1, 2),
        min_df=2,
        sublinear_tf=True
    )),
    ("clf", LinearSVC(
        C=1.0,
        random_state=RANDOM_STATE,
        dual="auto",
        max_iter=5000
    ))
])

advanced_xgb = Pipeline([
    ("tfidf", TfidfVectorizer(
        ngram_range=(1, 3),
        min_df=2,
        max_features=20000,
        sublinear_tf=True
    )),
    ("clf", XGBClassifier(
        n_estimators=300,
        max_depth=3,
        learning_rate=0.05,
        subsample=0.9,
        colsample_bytree=0.8,
        eval_metric="logloss",
        random_state=RANDOM_STATE,
        n_jobs=1
    ))
])

leaderboard_model = Pipeline([
    ("features", FeatureUnion([
        ("word", TfidfVectorizer(
            analyzer="word",
            ngram_range=(1, 3),
            min_df=1,
            max_features=15000,
            sublinear_tf=True
        )),
        ("char", TfidfVectorizer(
            analyzer="char_wb",
            ngram_range=(3, 5),
            min_df=2,
            max_features=20000,
            sublinear_tf=True
        ))
    ])),
    ("clf", LogisticRegression(
        C=2.0,
        class_weight="balanced",
        max_iter=1000,
        solver="liblinear",
        random_state=RANDOM_STATE
    ))
])

models = {
    "Baseline TF-IDF + Linear SVM": baseline_svm,
    "Advanced TF-IDF + XGBoost": advanced_xgb,
    "Leaderboard word+char TF-IDF + balanced Logistic Regression": leaderboard_model
}
```

```python
results = []

for name, model in models.items():
    print("\nTraining:", name)
    
    model.fit(X_train, y_train)
    pred = model.predict(X_val)
    
    score = f1_score(y_val, pred, average="macro")
    
    results.append({
        "model": name,
        "validation_macro_f1": score
    })
    
    print(f"Macro-F1: {score:.4f}")
    print(classification_report(y_val, pred, digits=4))

results_df = pd.DataFrame(results).sort_values(
    "validation_macro_f1",
    ascending=False
)

display(results_df)
```

```python
skf = StratifiedKFold(
    n_splits=3,
    shuffle=True,
    random_state=RANDOM_STATE
)

cv_scores = []

for tr_idx, val_idx in skf.split(X, y):
    leaderboard_model.fit(X.iloc[tr_idx], y.iloc[tr_idx])
    pred = leaderboard_model.predict(X.iloc[val_idx])
    
    score = f1_score(y.iloc[val_idx], pred, average="macro")
    cv_scores.append(score)

print("Final model CV scores:", [round(s, 4) for s in cv_scores])
print("Mean Macro-F1:", round(np.mean(cv_scores), 4))
print("Std:", round(np.std(cv_scores), 4))
```

```python
final_model = leaderboard_model

final_model.fit(X, y)

test_pred = final_model.predict(test_df["text"]).astype(int)

submission = pd.DataFrame({
    "id": test_df["id"],
    "label": test_pred
})

submission.to_csv("submission.csv", index=False)

print(submission.shape)
print(submission["label"].value_counts())

display(submission.head(10))
```
